```python
import pandas as pd

# Assuming your data is in a DataFrame named 'df' and the date column is named 'date'
def separate_date_columns(df):
    df = pd.read_csv('bangaluru data.csv')
    # Convert the 'date' column to datetime format (assuming 'YYYY/MM/DD' format)
    df['date'] = pd.to_datetime(df['date_ymd'], format='%Y/%m/%d')

    # Extract year, month, and day as separate columns
    df['year'] = df['date'].dt.year
    df['month'] = df['date'].dt.month
    df['day'] = df['date'].dt.day

    # Optionally, drop the original 'date' column if not needed
    df = df.drop('date_ymd', axis=1)  # Uncomment to drop the original 'date' column

    return df

# Apply the function to your DataFrame
df = separate_date_columns(df.copy())  # Operate on a copy to avoid modifying original data

print(df)  # View the DataFrame with separate year, month, day columns
```
